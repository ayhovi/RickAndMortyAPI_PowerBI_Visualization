```python
import requests
import pandas as pd
```

```python
# URL base de la API de Rick and Morty
base_url = "https://rickandmortyapi.com/api"
```

```python
# Función para obtener datos de la API
def fetch_data(endpoint):
    url = f"{base_url}/{endpoint}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        raise Exception(f"Error al obtener datos de {url}: {response.status_code}")
```

```python
# Obtener información de personajes
def get_characters():
    characters = []
    page = 1
    while True:
        data = fetch_data(f"character?page={page}")
        characters.extend(data['results'])
        if data['info']['next'] is None:
            break
        page += 1
    return characters
```

```python
# Obtener información de episodios
def get_episodes():
    episodes = []
    page = 1
    while True:
        data = fetch_data(f"episode?page={page}")
        episodes.extend(data['results'])
        if data['info']['next'] is None:
            break
        page += 1
    return episodes
```

```python
# Obtener información de ubicaciones
def get_locations():
    locations = []
    page = 1
    while True:
        data = fetch_data(f"location?page={page}")
        locations.extend(data['results'])
        if data['info']['next'] is None:
            break
        page += 1
    return locations
```
